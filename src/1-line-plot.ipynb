### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = x ** 2

# Validate data
if not np.issubdtype(x.dtype, np.number) or not np.issubdtype(y.dtype, np.number):
    raise TypeError("x and y must contain numeric values.")

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = x²', color='blue', linewidth=2)
plt.title("Line Plot of y = x²", fontsize=14)
plt.xlabel("x values")
plt.ylabel("y values (x squared)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
