### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x_line = np.linspace(0, 10, 100)
y_line = x_line ** 2

categories = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 9]
bar_colors = ['red', 'green', 'blue', 'orange']

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]
pie_colors = ['gold', 'skyblue', 'lightcoral', 'limegreen']

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line chart
axs[0, 0].plot(x_line, y_line, color='blue')
axs[0, 0].set_title("Line Chart: y = x²")
axs[0, 0].set_xlabel("x")
axs[0, 0].set_ylabel("y")
axs[0, 0].grid(True)

# Bar chart
axs[0, 1].bar(categories, bar_values, color=bar_colors)
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Category")
axs[0, 1].set_ylabel("Value")

# Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter, color='purple', s=80)
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("x")
axs[1, 0].set_ylabel("y")
axs[1, 0].grid(True)

# Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, colors=pie_colors, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title("Pie Chart")
axs[1, 1].axis('equal')  # Keep it circular

# Layout adjustment
plt.tight_layout()
plt.show()
```
