### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Line chart data
x_line = np.linspace(0, 10, 100)
y_line = x_line ** 2

# Bar chart data
categories = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 9]
bar_colors = ['red', 'green', 'blue', 'orange']

# Scatter plot data
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Pie chart data
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]
pie_colors = ['gold', 'skyblue', 'lightcoral', 'limegreen']

# Validate datasets
assert len(categories) == len(bar_values) == len(bar_colors)
assert len(x_scatter) == len(y_scatter)
if sum(pie_sizes) != 100:
    pie_sizes = [s * 100 / sum(pie_sizes) for s in pie_sizes]

# Subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].plot(x_line, y_line, color='blue')
axs[0, 0].set_title("Line Chart: y = x²")
axs[0, 0].grid(True)

axs[0, 1].bar(categories, bar_values, color=bar_colors)
axs[0, 1].set_title("Bar Chart")

axs[1, 0].scatter(x_scatter, y_scatter, color='purple')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].grid(True)

axs[1, 1].pie(pie_sizes, labels=pie_labels, colors=pie_colors, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title("Pie Chart")
axs[1, 1].axis('equal')

plt.tight_layout()
plt.show()
```
